```python
from pluto_survey_tools import QUESTIONNAIRE
import pluto_survey_tools.stats.frequencies as freq
```

```python
df1 = freq.score_count_df_keyed(questions=QUESTIONNAIRE.copy().questions, normalize=True)
df2 = df1.copy()
df2['score_x'] = df2['score_x'] + 15
df2['score_y'] = df2['score_y'] - 10
df2['score'] = df2['score'] + 5
```

```python
import altair as alt
import pandas as pd

x_min = min(df1['score_x'].min(), df2['score_x'].min())
x_max = max(df1['score_x'].max(), df2['score_x'].max())
y_min = min(df1['score_y'].min(), df2['score_y'].min())
y_max = max(df1['score_y'].max(), df2['score_y'].max())
domain_x = list(map(int, range(int(x_min), int(x_max) + 1)))
domain_y = list(reversed(list(map(int, range(int(y_min), int(y_max) + 1)))))

heatmap = alt.Chart(df1).mark_rect().encode(
    x=alt.X('score_x:O',
            title='Benefits',
            sort="ascending",
            axis=alt.Axis(labelExpr="datum.value % 2 ? null : datum.label"),
            scale=alt.Scale(domain=domain_x)
            ),
    y=alt.Y('score_y:O',
            title='Public value',
            sort="descending",
            scale=alt.Scale(domain=domain_y)
            ),
    color=alt.Color('count:Q',
                    title='Frequency',
                    scale=alt.Scale(scheme='viridis'),
                    legend=alt.Legend(orient='none', offset=0, legendX=-100, legendY=0)
                    ),
    tooltip=['score_x', 'score_y', 'count']
).properties(
    width=600,
    height=600
)
histogramX = alt.Chart(df1, width=600, height=100).mark_bar().encode(
    x=alt.X('score_x:O', axis=alt.Axis(labelExpr="datum.value % 2 ? null : datum.label")),
    y=alt.Y('count:Q', title='Frequency'),
)
histogramXDiff = alt.Chart(df2, width=600, height=100).mark_bar(color='red', opacity=0.05).encode(
    x=alt.X('score_x:O', axis=alt.Axis(labelExpr="datum.value % 2 ? null : datum.label")),
    y=alt.Y('count:Q', title='Frequency'),
)
histogramY = alt.Chart(df1, width=100, height=600).mark_bar().encode(
    x=alt.X('count:Q', title='Frequency'),
    y=alt.Y('score_y:O', sort="descending"),
)
histogramYDiff = alt.Chart(df2, width=100, height=600).mark_bar(color='red', opacity=0.05).encode(
    x=alt.X('count:Q', title='Frequency'),
    y=alt.Y('score_y:O', sort="descending"),
)
histX = histogramX + histogramXDiff
histY = histogramY + histogramYDiff
histX & (heatmap | histY)
```
